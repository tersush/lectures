```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем массив данных
dob_array = np.array([
    10, 23, 14, 16, 18, 20, 30, 24, 26, 28, 30, 32, 60, 36, 38, 50
])

# Построение графика
plt.figure(figsize=(10, 6))  # Устанавливаем размер графика
plt.plot(dob_array, marker='o', linestyle='--', color='b', label='Данные')  # Линия с маркерами

# Добавляем заголовок и подписи осей
plt.title('График значений из dob_array')
plt.xlabel('Индекс элемента')
plt.ylabel('Значение')

# Добавляем сетку и легенду
plt.grid(True)
plt.legend()

# Отображаем график
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Данные
dob_array = np.array([
    10, 23, 14, 16, 18, 20, 30, 24, 26, 28, 30, 32, 60, 36, 38, 50
])

# Создаем DataFrame с индексами
df = pd.DataFrame({
    "index": range(len(dob_array)),
    "value": dob_array
})

# Вариант 1: Линейный график
fig = px.line(
    df,
    x="index",
    y="value",
    title="График значений из dob_array",
    markers=True,  # Добавляем маркеры
    template="plotly_white"  # Чистый стиль
)

# Вариант 2: Столбчатая диаграмма (разкомментируйте)
# fig = px.bar(df, x="index", y="value", title="График значений из dob_array")

# Показываем график
fig.show()
```
